Create a Time Index

So we have all rides from January 1, 2001 to December 31, 2019. Let's do some statistics that group ride statistics by date. First we need to tell Pandas that the date column is in fact a date. So, we convert the column to a proper 'DateTime' type, and then set the index to this column.

This step takes a little bit of time!

The following picture show the desired output

image.png

```python
# hint, use to_datetime() and set_index()  modules
# your code here

# Convertir columna "release_date" a tipo "DateTime" y establecer como índice
movies = df.read_csv(url)
movies['date'] = df.to_datetime(movies['date'])
# Ordenar DataFrame por índice de tiempo
movies = movies.sort_index()
movies.set_index('date', inplace=True)

# Imprimir primeras filas del DataFrame con índice de tiempo
print(movies.head())
```

Sort the Dataframe by Date

```python
# hint, sort_values(by='date')
# your code here

# Convertir columna "release_date" a tipo "DateTime"
movies = df.read_csv(url)
movies['date'] = df.to_datetime(movies['date'])

# Ordenar DataFrame por columna "release_date"
movies = movies.sort_values(by='date')

# Imprimir primeras filas del DataFrame ordenado
print(movies.head())
```
